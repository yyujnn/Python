- python과 java의 차이  
python : 대화형 문법 (인터프리터) > 간단한 양의 데이터를 가져올 때  
java : 컴파일러 언어 > 많은 양의 데이터를 가져올 때
  

- 왜 python ?   
읽고 쓰기 편리함
비동기식 코딩방식 (인터프리터 언어) -> 한줄 한줄 처리해주는 장점 (대화형 언어)

  
- BIGDATA => 규모 Volume 속도 Velocity 다양성 Variety  + 정확성 Veracity 가치 Value  
디지털 환경에서 생성되는 수치, 문자, 이미지, 영상 데이터를 모두 포함하는 대규모 데이터


- 수집 데이터 형태  
정형   : 규격이 정리되어있음 ex) RDB, 스프레드시트, CSV  
반정형 : 규격은 정해져 있으나 데이터끼리 연산 불가 ex) XML, HTML, 지문  
비정형 : 규격 없이 자유로운 데이터 ex) SNS, 영상, CCTV, 해시태그   
목표 > 반정형, 비정형 데이터 -> 정형 데이터로 정제 (크롤링 통해)
  
  
- 웹 기초 지식   
Client -- 요청 (Request)-->  Server   
Client <-- 응답 (Response)--  Server 

### requests 라이브러리
- 서버에게 데이터를 요청해서 응답받기 위한 라이브러리
- 브라우저의 역할을 대신한다.

```python
# 라이브러리 로딩
import requests as req
```

```python
# naver 페이지 정보 받아오기

# 나 : req야 네이버 메인페이지 정보 좀 알아와줘
# req : 넵 근데 네이버가 어디에요?
# 나 : 'https://www.naver.com/'

res = req.get('https://www.naver.com/')
res

# Response [200] : 성공적으로 응답 받았다. 통신 성공!
# Response [400] : 클라이언트(요청)의 문제
# Response [500] : 서버(응답)의 문제
```

```python
# 요청한 url 주소의 페이지 보기
res.text 

# 결과가 문자열 형태의 데이터 -> 컴퓨터가 해석하기 어려움
```

### BeautifulSoup 라이브러리
- 컴퓨터가 이해할 수 있는 언어로 변환해주는 라이브러리
- bs 객체화

```python
# 넘겨받은 데이터가 text(문자열) 형태임
# 컴퓨터가 접근할 수 있는 html 형태로 변환 해주어야 함!

# 뷰티풀숩 라이브러리 로딩
from bs4 import BeautifulSoup as bs
```

```python
# BeautifulSoup 객체화 -> bs객체화 
# bs(페이지정보, '파싱방법')
soup = bs(res.text, 'lxml')
```

```python
# naver 페이지에서 '메일' 글자 수집
# select() : 원하는 요소 정보를 가져오는 기능 ex) soup.select('a.nav')
# 주의할 점 : 반환 값 무조건 리스트 형태이다. 
# 인덱싱이 필요하다
soup.select('a.nav')[0]
# 요소 = 태그 + 컨텐츠 <a>메일</a>
```

```python
# 요소에서 우리가 필요한 순수 컨텐츠만 추출 --> .text
soup.select('a.nav')[0].text
```

```python
# 메일 카페 블로그 글자 데이터 수집
tag = soup.select('a.nav')
# 메일만 추출
tag[0].text
```

```python
# 데이터의 길이를 알려주는 함수
len(tag)
```

```python
# 반복문을 활용해서 모든 데이터 추출
# range() : 범위를 생성해주는 함수 :
# range(시작할 값, 끝 값(포함x), 증감)  
for i in range(len(tag)):
    print(tag[i].text)
```

### 네이버 날씨에서 온도 데이터 수집
- 네이버 검색창에 '날씨' 검색 후 나오는 온도 데이터

```python
# 1. 네이버 날씨 url을 통해 날씨 페이지 정보 가져오기 (응답받기)
url = 'https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query=%EB%82%A0%EC%94%A8&oquery=%EB%82%A0%EC%94%A8&tqi=itYuTdprvh8sslqCQgwssssssw0-205564'
res = req.get(url)
# 2. bs 객체화 진행 -> bs(페이지정보, 'lxml')
soup= bs(res.text, 'lxml')
```

```python
# 3. 온도 데이터의 태그 위치를 선택
temp=soup.select('div.temperature_text>strong')
```

```python
# 4. 우리가 필요한 순수 데이터인 컨텐츠만 추출
temp[0].text
```

```python
# select_one() : 모든 태그 중 가장 상단에 있는 요소만 가져온다. (인덱스 0번)
# select : 모든 태그르 가져온다. 복수개, [리스트] 형태로 반환
# --> 그래서 .text를 해줄 때에는 반드시 인덱싱을 해주어야 함!
soup.select_one('div.temperature_text>strong').text
```
