### 특정 문자(열)에 접근하는 방법
- 인덱싱(Indexing)
- 슬라이싱(Slicing)

```python
#문자 인덱싱 , 슬라이싱
#s = 'This subject is python'
#b 접근 : 7 (앞에서부터), -15 (뒤에서부터)
#This 슬라이싱 : 0~ 4(3+1) 로.

#뒤에서 부터 셀 땐 -1 부터 센다.
```

```python
s = 'This subject is python'
```

```python
print(s[-15])
print(s[0 : 4])
```

```python
s = 'This subject is python'
s[0:4] # 시작 인덱스는 생략이 가능하기 때문에
s[:4]  # 생략하고 적을 수 있다.
s[5:12]
s[16:] # 마찬가지로 생략 가능
s[-6:]  #- 인덱스 부터 가져오기
```

```python
# 문장 처음부터 끝까지 슬라이싱( 첫~ 끝)
s[:]
```

```python
# s[시작인덱스 : 끝인덱스 : 증감값] 
# => 두 칸씩 이동하면서 데이터 접근.
s[::2]
```

```python
s[::-1]
# 문자열 전체 뒤집어서 가져오기
```

```python
day = '2023년 3월 10일의 날씨는 맑음입니다.'
print('날짜 : ', day[:12])     # ※ print 소괄호 안에서 , 로 연결 가능
print('날씨 : ', day[-6:-4])   # ※ 항상 끝 인덱스는 +1 을 해준다 
```

### 문자열 함수

```python
# day 문자열에서 '맑음' 인덱스 번호 확인하기
# find 함수 활용
a = day.find('맑음')   # 인덱스 번호 = 18 ('맑')
day[a:a+2]
```

```python
#find 함수 사용해 날짜 인덱스 번호 확인하기
# 날짜 전체 가져오려면 ~ 일,, 일의 인덱스를 찾고 거기까지 슬라이싱 할 것.
b = day.find('일')  
day[:b+1]
```

```python
# 함수 index 와 find 의 차이 비교 
# 문장안에서 없는 문자열을 찾을 때

day.index('e')  # 문자를 찾을 수 없다면 에러 발생
day.find('e')   # 찾는 문자 없을 때 -1 을 반환.
                # 에러를 발생하지 않는 상황으로 더 많이 사용된다.
```

```python
# len () : 문자열 개수 반환 (길이 반환)
len (day)  # 공백, !, ?, . 모두 포함 된 길이임
```

```python
# replace ('찾을 문자열', '변경할 문자열') : 특정 문자열을 찾아서 변경하는 함수!
# day : 맑음 -> 흐림                       ※ 그리고 다시 day 에 담아줄 수도 있다.
day = day.replace('맑음', '흐림')   

day
```

```python
#strip() : 양쪽 공백을 제거해주는 함수
day2 = ' 우리는 빅데이터 2교육실 입니다!!! '
day2
```

```python
day2.strip()
```

```python
# lstrip() : 왼쪽 공백을 제거하는 함수
# rstrip() : 오른쪽 공백을 제거하는 함수

day2.lstrip()
day2.rstrip()
```

```python
# count ('찾을 문자열') : '찾을 문자열'이 몇 개 있는지 개수 반환하는 함수
test = 'this subject is python!!!'
test.count('is')
test.count('s')
```

```python
#대문자를 - 소문자로 : lower()
#소문자를 - 대문자로 : upper()
test.upper()
```

```python
# split() : 문자열을 나눠주는 함수
test.split()          # 기준을 주지 않을 때 => 기본적으로 공백을 기준으로 나눠준다.
```

```python
# 특정 기호를 기준으로 문자열을 나누고 싶을 때  () 소괄호 안에 기호를 넣어주기
s2= '오늘을~ 즐거운~ 월요일 입니다~'
s2.split('~')
```

### 문자열 포매팅
- %기호 포매팅
- format 함수를 사용한 포매팅
- f 문자열 포매팅

#### %기호 포매팅

```python
# '%d' 를 이용해서 정수 대입해 보기
day 
a = 14
s = '내일은 3월 %d일 입니다.'% a
s 
```

```python
month = 3
day = 13
s = '오늘은 %d 월 %d일 입니다.' %(month,day)
s
```

```python
# 이렇게 % 포매팅에선 ( , ) 안에 자리를 잘 지정해 주는 게 중요
```

#### format 함수를 사용한 포매팅

```python
month = 3
day = 13
s = '오늘은 {} 월 {}일 입니다.' .format(month,day)
s
```

```python
# 내일은 3월 14일 입니다.
s = '내일은 3월 {}일 입니다.' .format(14)
s

# -> 먼저 대체할 값을 선언해주지 않고 중괄호 자리만 지정
```

```python
#인덱스 개념을 이용한 format 함수 포매팅
s = '오늘은 {0}월 {1}일 {1}요일 입니다.' .format(3,13)
s
```

#### f 문자열 포매팅

```python
month = 3
day = 13
s = f'오늘은 {month} 월 {day}일 입니다.' 
s
```

```python
# f 문자열 포매팅 이용 (자주 사용)

lang = "'python'"
s = f'Life is too short, You need {lang} '
print(s)
```

```python
# % 포매팅 이용  %s 값

lang = "'python'"
s = f'Life is too short, You need %s ' %lang
print(s)
```

```python
review = "[수료생 Real 후기] 후기입력 2022.10.13. 작성자 킹수. 드디어 \'인공지능 융합서비스 개발자과정\'을 수료했습니다!! 6개월이라는 길다면 길고 짧다면 짧은 기간 동안 많은 노력을 통해 받아낸 수료인 만큼 마음이 벅차네요. 그치만 정말 재밌고 즐거운 6개월이었습니다. 많은 것을 배웠고, 또 자신감도 얻은 기간이었다고 생각합니다. 이 후기를 본 많은 분들이 저를 따라서 수료를 목표로 도전하셨으면 좋겠습니다!!! 사랑해요 여러분"
```

```python
start = review.find('2')
end = review.find('3')
print(start)
print(end)
```

```python
day = review[start:end+1]     # 이렇게 index no 대신 저장해둔 변수를 사용할 수 있다

print(f'후기가 업로드 된 일자는 {day} 입니다.')
```
