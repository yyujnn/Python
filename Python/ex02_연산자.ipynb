### 비교 연산자

```python
# True, False로 결과가 출력되는 연산자
# >, >=. <, <=, == (같다), !=(같지않다)

a = 3
b = 7

print(a > b) # False
print(a <= b) # True
print(a == b) # False
print(a != b) # True
```

### 논리 연산자

```python
# not, and, or
# not : 논리를 뒤집어 버리는 연산자 True -> False, False -> True
a, b = 5, 10
not a < b  # True --> False
```

```python
# and , or
# and : 두 개의 논리를 연결해서 확인. 두 논리 모두 true => true
# or : 한 개의 논리만 참일 경우에도 true => true

print(3 > 5 and 10==10)  #False
print(3 >5 or 10==10)  #True
```

```python
'2교육실' not in str1  # False  'str1 이란 거에 2교육실 이라는 말이 없지?'
```

### 삼항 연산자
- True 일때 실행 문장     if    조건식   else     False 일때 실행문장


```python
# 60점 이상이면 합격, 미만이면 불합격 출력해보기
score = 50
'합격' if score > 60 else '불합격'
```
