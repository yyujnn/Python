### 리스트(list)
- 파이썬 자료구조 중 하나
- 순서가 있고, 수정 가능한 자료구조
- 대괄호([ ])로 작성이 되고, 리스트 내부의 값은 콤마 (,) 로 구분
- 하나의 리스트에 다양한 자료형을 포함
- 추가, 수정, 삭제가 가능하다.

#### 리스트 선언

```python
# 리스트명 = [요소1, 요소2, 요소3, ...]
a = []  # 빈 리스트 선언
b = [1, 2, 3]   # 숫자 요소를 담은 리스트
c = ['우민철', '조인성', '고주진']   # 문자열 리스트
d = [1, 2, '이경진']  # 숫자, 문자열 요소를 담은 리스트
e = [1, 2, '김윤', ['문동현', '이상원']]   #리스트를 담은 리스트

# [] = 단일 리스트
# [[]] = 이중 리스트
```

```python
print(a)
print(b)
print(c)
print(d)
print(e)
```

#### 리스트 데이터 접근하기 : 인덱싱 & 슬라이싱을 사용!

```python
# 인덱싱 : 리스트명[인덱스]
lst1 = [1, 3, 5, 7, 9]
# 인덱스 번호 0번 데이터 접근
lst1[0]  # 결과 값 : 1
# '7' 이라는 요소에 접근하기
lst1[3]
```

```python
# 이중 리스트

list2 = [1, 2, 3, ['a', 'b', 'c']]
temp = list2[3]

print(temp)

print(list2[3][1])   #이중 리스트에서 원하는 요소에 접근할때는 대괄호도 2개가 필요하다.
```

```python
# len() : 길이를 알려주는 함수를 이용해서 요소의 개수를 확인하기
print(len(list2))
```

- 슬라이싱

```python
# 리스트명 [시작인덱스 : 끝인덱스]
# 시작인덱스 <= 인덱스 범위 < 끝 인덱스  ==> 그래서 끝인덱스 +1 을 해줘야 했다.

lst2 = ['델로스', '누네띠네', '엄마손파이', '후레쉬파이','촉촉한초코칩']

# '누네띠네' ~ '후레쉬파이' 데이터 접근하기

lst2[1:4]    # ◎ 출력 결과도 리스트로 나온다!
```

```python
# '델로스' ~ '촉촉한초코칩' 까지 모두 접근해보기

lst2[ : ]    # 첫 인덱스와 끝 인덱스 모두 생략해서 : 만 적어주세요.
```

```python
# 리스트명 [시작인덱스 : 끝인덱스 : 증감값]

lst2[1::3]     # 리스트 1번 부터 끝번 까지 3씩 커지면서 접근
```

```python
# 이중리스트를 선언하고 데이터 접근해보기
# 이름, 나이, 혈액형 정보 기반의 이중 리스트를 선언해 보자!

lst3 = [['김동현', 20, 'B'], ['선동원', 36, 'AB'], ['이경진', 18, 'B'],['방은혜', 26, 'B'], ['김수진', 26, 'B']] 
```

```python
lst3
```

```python
# 김수진씨의 나이, 혈액형 데이터 접근해서 출력하기
# 수진씨의 나이는 ()이고, 혈액형은()입니다.

lst3[4]
'수진씨의 나이는 {}이고, 혈액형은 {}입니다.'.format(lst3[4][1],lst3[4][2])
f'수진씨의 나이는 {lst3[4][1]}이고, 혈액형은 {lst3[4][2]}입니다.'
```

```python
# 선동원씨의 정보와 방은혜씨의 정보를 리스트로 가져오기
lst3[1::2]  #슬라이싱을 사용해서 둘의 인덱스에 접근 가능.
```

### 튜플 (tuple)
- 파이썬 자료구조 중 하나
- 순서가 있는 자료(인덱싱, 슬라이싱 기능)
- 소괄호()로 작성, 튜플 내부의 값은 콤마 , 로 구분
- 하나의 튜플에는 다양한 자료형을 포함
- 데이터의 추가, 수정, 삭제 ==> 불가능
- 언제 사용할까? 프로그램이 실행되는 동안, 요소값이 항상 변하지 않기를 바라거나, 값이 바뀔까 걱정하고 싶지 않을 경우에 사용

```python
a = () # 빈 튜플
b = (1, 2, 3)  #숫자형 데이터를 담은 튜플
c = ('고주진', '이민혁', '조인성')  # 문자열 데이터를 담은 튜플
d = (1, 2, '김수진', '우민철')  # 숫자 문자열 모두 담은 튜플
e = (('빅데이터 3차', '오선영', 25), ('빅데이터 3차', '맹성영', 15)) #이중 튜플

```

```python
print(a)
print(b)
print(c)
print(d)
print(e)
```

#### 튜플 데이터 추가, 수정, 삭제하기  ==> 불가능하다!!

```python
# 튜플 e 에 접근해서 '오선영' -> '김수진' 변경해보기
# e[0][1] = '김수진'   # ==> 오류!!
# 튜플은 데이터를  변경할 수 없다!
```

```python
lst_f = ['딸기','바나나', '복숭아', '체리', '사과']
```

```python
fruits = input('검색할 문자를 입력하세요 >> ')

if fruits in lst_f :
   
    
    print(f'{fruits}는 리스트 {lst_f.index(fruits)}번째 인덱스에 들어있습니다.')
else :
    print(f'{fruits}는 리스트에 들어있지 않습니다.')
#   리스트에서 쓸 수 있는 찾는 함수는 ※ index 만 있다.
```

```python
a = 'testpythontestpythontestpythontestaaaaa'
a.count('t')
```

```python
s = 'Hi, My name is Juhee'
w = input('검색할 문자를 입력하세요 >> ')

if w in s :
    print(f'{w}는 {s.count(w)}번 들어 있습니다')
else :
    print(f'{w}는 문자열에 들어있지 않습니다.')
```
