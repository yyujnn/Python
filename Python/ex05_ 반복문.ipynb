### while문 
- 반복 횟수가 명확하지 않을 때

```python
# 파이썬 최고!!!  를 13 번 출력
num = 1

while num <=13 :
    print('파이썬 최고!!!')
    num+=1
```

```python
# break 사용
while True :
    print('무한루프')
    break
```

```python
# continue

num = 0
while num < 10 :
    num+=1  # num = num + 1
#    continue를 만나는 순간 다시 조건문으로 간다. if 만족 x -> print!
    if num%2 == 0 :
        continue
    print(num) # else 없이 자리에 맞춰서 사용
#     --> 홀수 값만 출력 됨
```

```python
while True :
    num1 = int(input('첫 번째 정수 입력 >> '))      # 계속 반복해서 받아야 하니까
    num2 = int(input('두 번째 정수 입력 >> '))      # while 문 안으로 넣어줘야함
   
    if num1==0 and num2==0 :
        print('프로그램이 종료되었습니다')
        break
    else :
        print(f'두 정수의 합 : {num1+num2}')
```

```python
num=1
now = int(input('현재 몸무게 : '))
goal = int(input('목표 몸무게 : '))

while True :
    lose = int(input(f'{num}주차 감량 몸무게 : '))
    now-=lose # 현재 몸무게에서 빠지는 값을 누적 해서 빼기
    num+=1
    if now<=goal :
        print(f'{now} 달성! 축하합니다')
        break
        
```

### for 문 
- 반복 횟수가 정해져 있을 때

```python
# 리스트를 사용할 것

lunch = ['김치볶음밥', '수육', '라면']
lunch
```

```python
for i in lunch :  # for i in ['김치볶음밥', '수육', '라면']
    print(i)    ## i 에다가 순차적으로 리스트 요소를 하나씩 담은 것
```

```python
len(lunch)
```

```python
# 문자열 반복
hi = '안녕하세요 여러분~~^^'

for i in hi :
    print(i, end = '')  # 출력문의 끝 속성
                        # 개행 \n을''로 대체해줘서 옆으로 적을 수 있었다
#함수의 속성 확인하기 : shift + tab
```

#### for : range() 
- 필요한 만큼의 숫자를 만들어 내는 유용한 기능
- range(시작할 숫자, 종료할 숫자, 증감량)

```python
for i in range(1, len(lunch)):  # 1 부터 3까지의 인덱스 = i 
    print(i, lunch[i]) # 인덱싱 해주는 개념. 인덱싱 해서 값을 가져오는 것,
```

```python
for i in range(1, 50, 3) : #1~49 까지의 숫자를 3씩 건너뛰면서 i 에 담는다.
    print(i, end=' ')
```

#### min, max 함수
- 선언한 자료구조(리스트, 튜플)에서 가장 큰 값, 가장 작은 값을 찾아주는 함수

```python
lst1 = [4, 5, 2, 30, 91, 7, 27]
```

```python
# lst1에서 가장 작은 수 구하기
min(lst1)
```

```python
# lst1에서 가장 큰 수 구하기
max(lst1)
```

```python
# 1~100 사이의 숫자중 3의 배수의 합을 구하기
result = 0 # 값을 담을 곳
for i in range(0, 100, 3):
    result+=i
print(result)
```

```python
# if 조건문 사용
result = 0
for i in range(1, 100) : # 1~100 범위 중에 변수i에 하나씩 담아서 실행문장 실행
    if i%3 ==0:
        result+=i
print(result)
```

### pickle
- 파이썬에서 리스트, 튜플 등의 자료구조 형태를 유지하면서 파일로 저장, 불러올 때 사용 하는 모듈

- 불러올 때 : rb(read binary), 저장할 때: wb (write binary)

### 멜론차트 Top100 활용해서 데이터를 분석해보고 나만의 차트 만들기

```python
import pickle
```

```python
# with문 사용해서 데이터 불러오기
with open('music_100.pkl', 'rb') as f:
    music = pickle.load(f)
```

```python
# 멜론차트 데이터를 불러와서 데이터 행, 열 정보 확인하기
music[:5]
# 순위 정보 , 노래제목, 가수명  이 하나로 묶인 리스트
```

```python
key = input('제목 중 찾고 싶은 키워드를 입력하세요 >> ')
s_cnt = 0
song_list=[] #찾은 노래 '순위정보'와 '제목' 담을 리스트

for i in music:
    if key in i[1] :  # i = [  ,  ,  ] i 번째 리스트를 가져온 것임
        s_cnt +=1
        song_list.append(i[:2])  # 0번(순위) ~ 1번(제목) 정보를 슬라이싱으로 append

print(f'\'{key}\'으로 검색한 노래 수 : {s_cnt}곡')
print(f'순위와 노래 제목 : {song_list}')
```
