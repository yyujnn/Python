### 딕셔너리란?
- Key와 Value를 한 쌍으로 갖는 자료형
- Key = 불변, Value = 불변/가변 한쌍으로 맵핑된 순서가 업는 집합
- 인덱싱, 슬라이싱이 불가능

### 딕셔너리 선언

```python
a = {} # 빈 딕셔너리 
b = {'name' : 'gyoung jin'} #이름정보 (문자열)을 담은 딕셔너리
c = {1 : 'gyoung jin', 2 : 20}  #하나의 딕셔너리에 다양한 자료형 담을 수 있다.
d = {'name' : ['이수', '상원', '민규']}  # value 값으로 리스트(자료구조)를 담을 수 있다.
# 뭔갈 명확히 정의 하고 싶을 때 딕셔너리를 사용
```

```python
print(a)
print(b)
print(c)
print(d)
```

### 딕셔너리 관련 함수
- 데이터에 접근하는 방법들

#### 딕셔너리 키값만 가져오기

```python
a.keys()
list(a.keys())
list(a.keys())[0]

# 만약 name 이라는 키에 접근하고 싶을때 (키 값으로 인덱싱 하겠다.)
# 리스트로 *형변환* 하여 키 값들을 인덱싱으로 접근할 수 있다.
```

#### 딕셔너리 value 값만 가져오기

```python
a.values()
list(a.values())[3]    # 생일에만 접근
```

#### 딕셔너리 key, value 모두 가져오기  : ※ items()

```python
a.items()     # 튜플 형태로 값을 출력해준다.
```

```python
list(a.items())
```

```python
for i in a.items():  # .items() = 튜플 + 튜플 + ... 인 전체 튜플을 가져오는 것, 
                    # i 를 가져오는 것은 = 큰 튜플 내의 하나의 작은 튜플단위로 가져온다는 것.
    print(i)           # 패킹된 형태로 출력 (하나니까)
```

```python
for key, value in a.items() :
    print(key, value)   # unpacking
```

```python
score_dict = {'이름':['성춘향', '이몽룡', '김방자', '최향단', '변학도'], 
             'Python':[90, 100, 85, 90, 80],
             'Java':[85, 100, 80, 95, 85],
             'html/css':[75, 70, 90, 80, 90]}
```

```python
list(score_dict.values())[1]
```

```python
a, b, c, d, e = list(score_dict.values())[1] 
print(a, b, c, d, e)
```

```python
py_avg = (a+b+c+d+e)/5
print(py_avg)
```

```python
# 총합을 구하는 함수 : sum()
# 1. python 과목에 접근해서 결과 가져오기
score_dict.keys()
score_dict['Python']   # Python키의 값들을 가져오기

# 2. 1번 합을 사람 수만큼 나누기, len()

sum(score_dict['Python'])/len(score_dict['Python'])
```

```python
# 각 과목의 평균을 for 문을 활용해서 구하기
for i in score_dict.keys() :
    print(i)
```

```python
# 이름을 빼고 과목만 가져오기 위해서 리스트로 형변환(인덱싱, 슬라이싱)
avg = 0
for i in list(score_dict.keys())[1:] :
    avg = sum(score_dict[i])/len(score_dict[i])
    print(f'{i}평균 : {avg}')
```
