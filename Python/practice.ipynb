```python
def op_cal (num1,num2) :
    if op == '+':
        result = num1+num2
        return result
    elif op == '-':
        result = num1-num2
        return result
    elif op == '*' :
        result = num1*num2
        return result
    elif op == '/' :
        result = "%0.6f"% float(num1/num2)
        return result
```

```python
print('====사칙연산(+,-,*,/)====')
num1 = int(input('첫 번째 정수 입력 : '))
num2 = int(input('두 번째 정수 입력 : '))
op = input('연산자 입력 : ')
result = op_cal(num1, num2)
print(f'연산결과 : {result}')
```
