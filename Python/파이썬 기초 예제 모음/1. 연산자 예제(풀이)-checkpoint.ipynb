### 1. 다음 코드를 완성하여 다음과 같은 결과를 출력


```python
# 코드 입력
num1 = 23
num2 = 3

print(f"더하기 결과 : {num1 + num2}")
print(f"빼기 결과 : {num1 - num2}")
print(f"곱하기 결과 : {num1 * num2}")
print(f"나누기 결과 : {num1 / num2}")
```

### 2. 다음 코드에서 변수 num1과 num2를 키보드로 입력 받은 뒤 결과를 계산


```python
# 코드 입력
num1 = int(input("정수 입력 >> "))
num2 = int(input("정수 입력 >> "))

print(f"더하기 결과 : {num1 + num2}")
print(f"빼기 결과 : {num1 - num2}")
print(f"곱하기 결과 : {num1 * num2}")
print(f"나누기 결과 : {num1 / num2}")
```

### 3. Python, 머신러닝, 딥러닝 점수를 키보드로 입력 받아 합계와 평균을 출력


```python
# 코드 입력
py = int(input("python 점수 입력 >> "))
ml = int(input("머신러닝 점수 입력 >> "))
dl = int(input("딥러닝 점수 입력 >> "))

# 1. 즉시 출력
print(f"합계 : {py + ml + dl}")
print(f"평균 : {(py + ml + dl) / 3}")

# 2. 변수에 저장 후 출력
sum_pmd = py + ml + dl
print(f"합계 : {sum_pmd}")
print(f"평균 : {sum_pmd / 3}")
```

### 4. 초를 입력 받아 "OO시간 OO분 OO초" 형태로 출력


```python
# 코드 입력
time = int(input("시간 입력 >> "))
hour = time // 3600
minute = time % 3600 // 60
second = time % 3600 % 60
print(f"{hour}시간 {minute}분 {second}초")
```

### 5. 키보드로 정수를 입력 받아 홀수/짝수 판별


```python
# 코드 입력
num = int(input("정수 입력 >> "))

# 1. 즉시 출력
print(f"{num}는(은) 짝수입니다.") if num % 2 == 0 else print(f"{num}는(은) 홀수입니다.")

# 2. 변수에 저장 후 출력
rs = "짝수" if num % 2 == 0 else "홀수"
print(f"{num}는(은) {rs}입니다.")
```

### 6. 두 개의 정수를 입력 받아 큰 수에서 작은 수를 뺀 결괏값 출력


```python
# 코드 입력
num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("두 번째 정수 입력 >> "))

# 1. 즉시 출력
print(f"두 수의 차 : {num1 - num2}") if num1 > num2 else print(f"두 수의 차 : {num2 - num1}")

# 2. 변수에 저장 후 출력
rs = num1 - num2 if num1 > num2 else num2 - num1
print(f"두 수의 차 : {rs}")
```
