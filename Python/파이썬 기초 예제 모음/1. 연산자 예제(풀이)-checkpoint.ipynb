### 2.    num1 num2      


```python
#  
num1 = int(input("  >> "))
num2 = int(input("  >> "))

print(f"  : {num1 + num2}")
print(f"  : {num1 - num2}")
print(f"  : {num1 * num2}")
print(f"  : {num1 / num2}")
```

### 3. Python, ,        


```python
#  
py = int(input("python   >> "))
ml = int(input("   >> "))
dl = int(input("   >> "))

# 1.  
print(f" : {py + ml + dl}")
print(f" : {(py + ml + dl) / 3}")

# 2.    
sum_pmd = py + ml + dl
print(f" : {sum_pmd}")
print(f" : {sum_pmd / 3}")
```

### 4.    "OO OO OO"  


```python
#  
time = int(input("  >> "))
hour = time // 3600
minute = time % 3600 // 60
second = time % 3600 % 60
print(f"{hour} {minute} {second}")
```

### 5.     / 


```python
#  
num = int(input("  >> "))

# 1.  
print(f"{num}() .") if num % 2 == 0 else print(f"{num}() .")

# 2.    
rs = "" if num % 2 == 0 else ""
print(f"{num}() {rs}.")
```

### 6.            


```python
#  
num1 = int(input("    >> "))
num2 = int(input("    >> "))

# 1.  
print(f"   : {num1 - num2}") if num1 > num2 else print(f"   : {num2 - num1}")

# 2.    
rs = num1 - num2 if num1 > num2 else num2 - num1
print(f"   : {rs}")
```
