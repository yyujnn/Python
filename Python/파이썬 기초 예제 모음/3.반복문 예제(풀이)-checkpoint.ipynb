# while문 예제

### 1. 이름을 입력 받고 다음과 같이 출력을 반복
- 단, 입력 받은 이름이 "이수환" 이라면 해당 출력문을 출력후 반복을 종료


```python
# 코드 입력
while True :
    name = input("이름 입력 >> ")
    
    if name == "이수환" :
        print(f"{name} 멋지다!")
        break
        
    print(f"{name} 최고!")
```

### 2. 두 개의 정수를 입력 받아서 더하는 코드를 작성
- 단, 두 개의 정수가 0이 들어올 때 까지 반복


```python
# 코드 입력
while True :
    num1 = int(input("첫 번째 정수 입력 >> "))
    num2 = int(input("두 번째 정수 입력 >> "))
    
    if num1 == 0 and num2 == 0 :
        break
    
    print(f"두 정수의 합 : {num1 + num2}")
          
print("프로그램이 종료되었습니다.")
```

### 3. 랜덤으로 1부터 50사이의 숫자를 뽑으면 뽑은 숫자를 맞추는 Up, Down 게임


```python
# 코드 입력

# 랜덤 라이브러리 불러오기
import random

#1~50까지 (끝값 포함O)
comNum = random.randint(1, 50)

while True :

    num = int(input("숫자를 입력하세요 >> "))

    if comNum == num :
        print("정답을 맞추셨습니다.")
        break
        
    elif comNum > num :
        print(f"{num}보다 큰 수 입니다.")
        
    elif comNum < num :
        print(f"{num}보다 작은 수 입니다.")
```

# for문 예제

### 1. 5명에 대한 시험 점수가 리스트에 담겨있다. 각 점수의 합격/불합격을 판별
- 60점 이상 합격

```python
# 코드 입력
score_list = [90, 45, 70, 60, 55]

num = 1

for i in score_list :
    
    if i >= 60 :
        print(f"{num}번 학생은 합격입니다.")
        
    else :
        print(f"{num}번 학생은 불합격입니다.")
    num += 1
```

### 2. for문을 사용하여 원하는 범위의 수 출력

2. 23부터 39까지 가로 출력

### 3. list2에 들어있는 요소 중 가장 큰 수를 찾아 출력


```python
# 코드 입력

list2 = [4, 5, 2, 1, 99, 15, 2, 7, 27]

max_num = list2[0]

for i in list2 :
    
    if i > max_num :
        max_num = i

# 살아남은 가장 큰 수 출력
print(max_num)
```

### 4. list2에 들어있는 요소 중 가장 작은 수를 찾아 출력


```python
# 코드 입력

list2 = [4,5,2,1,99,15,2,7,27]

min_num = list2[0]

for i in list2 :
    
    if i < min_num :
        min_num = i
        
# 살아남은 가장 작은 수 출력
print(min_num)
```

### 5. 1부터 100사이의 숫자 중 3의 배수인 값들의 합 출력
- 정답 : 1683

```python
# 코드 입력

# range함수 활용

total = 0

for i in range(0, 101, 3) :
    total += i
    
print(total)

# 조건문 활용

total = 0

for i in range(1, 101) :
    
    if i % 3 == 0 :
        total += i
        
print(total)
```

### 8. 반복문을 사용하여 다음과 같이 별(*)을 표시하는 프로그램 작성


```python
# 코드 입력

# for문 사용
for i in range(1, 7, 1) :
    print("*" * i)

# 개행
print()
    
# while문 사용
i = 0
while i <= 5 :
    i += 1
    print("*" * i)
```

### 12. 세 개의 정수를 입력 받아 첫 번째 정수부터 두 번째 정수까지 범위에서<br>입력된 배수의 합을 출력


```python
# 코드 입력
num1 = int(input("첫 번째 정수 입력 >> "))
num2 = int(input("두 번째 정수 입력 >> "))
num3 = int(input("배수 입력 >> "))

sum_num = 0

for i in range(num1, num2+1, 1) :
    
    if i % num3 == 0 :
        sum_num += i
        
print(f"{num3}의 배수의 합 :", sum_num)
```

### 13. 다음 암호문을 올바르게 해석하려면 거꾸로 읽어야 한다.<br>암호문을 해석하기 위한 반복문 작성
- len() 함수 이용


```python
# 코드 입력

# 첫 번째 방법
s = ".nohtyp deen uoY ,trohs oot si efiL"

for i in range(len(s)-1, -1, -1) :
    print(s[i], end = "")

# 개행    
print()

# 두 번째 방법
s = ".nohtyp deen uoY ,trohs oot si efiL"

for i in range(-1, -len(s)-1, -1) :
    print(s[i], end = "")
```

### 16. list1에서 짝수/홀수의 개수를 출력
- 단, list1의 값이 추가/수정 되어도 출력되게끔 코드 작성


```python
# 코드 입력
list1 = [1, 30, 7, 9, 17, 3, 5, 23, 44]

odd = 0
even = 0

for i in list1 :
    
    if i % 2 == 0 :
        even += 1
    else :
        odd += 1
        
print(f"짝수 {even}개, 홀수 {odd}개")
```

### 17. 다음 score_list에 들어있는 데이터를 이용해 아래와 같이 성적표를 만드는 프로그램 작성
- 단, 학생의 숫자가 늘어나도 동작할 수 있도록 고려


```python
# 코드 입력

score_list = [["박매일", 89, 78, 99],
                       ["김미희", 67, 88, 77],
                       ["강예진", 100, 55, 96]]

print("=" * 50)
print("이름\t", "Python\t", "Java\t", "DB\t", "평균\t", "학점")  # 출력결과 1번째 줄
print("=" * 50)

python, java, db = 0, 0, 0

for i in score_list :  # i = score_list[0], score_list[1], score_list[2]
    
    avg = ( i[1] + i[2] + i[3] ) / 3
    avg_P, avg_J, avg_D = 0, 0, 0
    
    if avg >= 90 :
        grade = "A"
        
    elif avg >= 80 :
        grade = "B"
        
    elif avg >= 70 :
        grade = "C"
        
    elif avg >= 60 :
        grade = "D"
        
    else :
        grade = "F"
    
    print(i[0], "\t", i[1], "\t", i[2], "\t", i[3], "\t", int(avg), "\t", grade)   # 출력결과 3~5번째 줄 
    
    python += i[1]
    java += i[2]
    db += i[3]
    
avg_P = python / len(score_list)
avg_J = java / len(score_list)
avg_D = db / len(score_list)

print("=" * 50)
print("평균", "\t", int(avg_P), "\t", int(avg_J), "\t", int(avg_D))  # 출력결과 마지막 줄
```
