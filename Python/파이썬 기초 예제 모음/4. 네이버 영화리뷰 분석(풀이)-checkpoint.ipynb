# pickle 라이브러리를 활용한 for문 예제

### pickle 라이브러리란?
- 텍스트 상태의 데이터가 아닌 리스트 혹은 클래스와 같은 자료형 객체를 저장하거나 불러오는 방법


## 잠깐! 데이터 로드 하기 전 확인
- 파일 다운로드 유무 (movie_review.pkl)
- 파일 경로 (현재 디렉토리와 같은 위치에 저장할 것)

### pickle 라이브러리를 활용하여 데이터 로드


```python
# 코드 작성
import pickle

# <pickle 파일 오픈 모드>
# 'wb' : write binary 모드(읽고 쓰기 가능)
# 'rb' : read binary 모드

# binary : 컴퓨터가 처리하는 파일 형식(사람이 읽기 힘듦)

with open('movie_review.pkl', 'rb') as f :
    my = pickle.load(f)
```

#### Step!
1. 전체 데이터의 개수 확인
2. 리뷰만 출력
3. 리뷰 길이 최댓값 / 리뷰의 내용
4. 리뷰 길이 최솟값 / 리뷰의 내용
5. 리뷰 길이 평균값
6. 긍정리뷰 / 부정리뷰 개수
7. 특정 문자가 포함된 리뷰의 개수

### 1. 전체 데이터 개수 확인


```python
# 코드 입력
len(my)
```

### 2. 리뷰만 출력


```python
# 코드 입력
for i in my :
    print(i[0])
```

### 3. 리뷰 길이 최댓값 / 리뷰의 내용


```python
# 코드 입력
max_len = len(my[0][0])
max_content = ''

for i in my :
    
    r_len = len(i[0])
    
    if max_len < r_len :
        max_len = r_len
        max_content = i[0]
        
print(f"리뷰 최대길이는 '{max_len}'입니다.")
print(max_content)
```

### 4. 리뷰 길이 최솟값 / 리뷰의 내용


```python
# 코드 입력
min_len=len(my[0][0])
min_content = ''

for i in my :
    
    r_len = len(i[0])
    
    if min_len > r_len :
        min_len = r_len
        min_content = i[0]
        
print(f"리뷰 최소길이는 '{min_len}'입니다.")
print(min_content)
```

### 5. 리뷰 길이 평균값


```python
# 코드 입력
total = 0

for i in my :
    total += len(i[0])
    
print(f"리뷰의 평균 길이는 '{total / len(my)}'입니다.")
```

### 6. 긍정리뷰 / 부정리뷰 개수
- 1은 긍정 / 0은 부정


```python
# 코드 입력
# 1은 긍정 / 0은 부정

# [1] 복합대입 연산자 활용
neg_cnt = 0
pos_cnt = 0

for i in my :
    
    if i[1] == 0 :
        neg_cnt += 1
        
    else :
        pos_cnt += 1
    
print(f"긍정 '{pos_cnt}'개")
print(f"부정 '{neg_cnt}'개")

# 개행
print()

# [2] 리스트 활용
re_list = []

for i in my :
    re_list.append(i[1])
    
neg = re_list.count(0)
pos = re_list.count(1)

print(f"긍정 '{pos}'개")
print(f"부정 '{neg}'개")
```

### 7. 특정 문자가 포함된 리뷰 개수


```python
# 코드 입력
cnt = 0
data = input("검색할 데이터 입력 >> ")

for i in my :
    
    if data in i[0]:
        cnt += 1
        
print(f"'{data}'가 포함된 리뷰는 {cnt}개 입니다. ")
```
