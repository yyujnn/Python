### 데이터를 시각화하는 라이브러리

#### 시각화 라이브러리(모듈) 불러오기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### line plot : 선 그래프

```python
# 선 그래프 키워드 : plot()
y = [2,4,6,8]
plt.plot(y)
plt.show()
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y)
plt.show()
```

#### 그래프 스타일 지정

```python
# 선 스타일
plt.plot(y, ls = '--')
plt.show()
```

```python
# 선 스타일
plt.plot(y, ls = ':', marker = 'D')
plt.show()
```

```python
# 선 두께 : line width --> lw
plt.plot(y, lw = 5)
plt.show()
```

```python
# 선색상 : line color -->  c
plt.plot(y,marker='*',lw=3, c='g')
plt.show()
```

```python
# marker의 컬러 지정하기
plt.plot(y, marker='D', c='g', mfc='y')
plt.show()
```

```python
x = np.arange(7)
y=[1,4,5,8,9,5,3]
plt.plot(x,y,ls=':', c='c', marker='s', mfc='y')
plt.show()
```

```python
# 그래프 사이즈 지정
plt.figure(figsize = (5,3))
plt.plot(x,y,ls=':', c='r', marker='s', mfc='b', ms=5, mec='g')
```

#### 그래프 옵션 지정

```python
# 그림 범위 지정
plt.plot(x,y,ls=':', c='r', marker='s', mfc='b', ms=5, mec='g')
plt.xlim(-2,10)  # x 축의 범위 제한
plt.ylim(2,8)  # y 축의 범위 제한
plt.show()

# 만약 2~8 사이의 값을 보고 싶을 때(zoomin) 
```

```python
# ticks : x, y 축의 표시값 설정
plt.plot(x,y,ls='--', c='r', marker='s', mfc='b', ms=5, mec='g')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
```

```python
# 눈금선 표시 : grid()
plt.plot(x,y,ls='-', c='b', marker='s', mfc='y', ms=5, mec='r')
plt.grid() # grid 함수 호출
plt.show()
```

```python
# 여러개의 그래프 출력하기
x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]
plt.plot(x,label='x')
plt.plot(y,label='y')
plt.plot(z,label='z')
plt.legend(loc='best') # 범례표시 기본값 : best
plt.xlabel('x') # x축 이름
plt.ylabel('y') # y축 이름
plt.title('my graph') # 그래프 제목
plt.show() # x, y, z 모두 출력
```

#### 요일별 교통사고 시각화 실습

##### 교통사고 사망자 수 count

```python
# value_count() : 특정 컬럼의 값들의 개수를 각각 출력
# Series(ex 요일 컬럼) 여러개 모여서 DataFrame
# 하나의 컬럼안에(=시리즈에서) 있는 유니크(중복 없는) 값들과 그 값들의 개수를 각각 출력
# 시리즈에서만 사용 가능하다 !!

data['사망자수'].value_counts() 
```

```python
# data.value_counts() # 데이터 프레임 value_counts 구체적이지 않음
```

- 위 결과로 알 수 있는 점 : 우리의 데이터는 사망자가 1인인 사고가 교통사고의 대부분을 차지하고 있음

```python
display(data['사망자수'])
data['사망자수'].count()
```

#### 요일별 사고 건수  count
- 요일 사고건수 확인
- 요일 순서대로(월~일) 출력

```python
# 1. 요일 컬럼 접근
data['요일']
```

```python
# 2. 중복 없이 값 확인(월~일)
# unique() : 중복 제거
data['요일'].unique()
```

```python
# 3. 요일별 사고 건수 확인 > Series에서 사용
y = data['요일'].value_counts()
y
```

```python
y.sort_values() # values로 정렬
y.sort_index() # 가나다 순
```

```python
# 4. 요일 순서대로 변경
# 행과 열로 이루어진 2차원 형태로 출력하기 위해 대괄호 2개를 사용하여 차원을 
y=y[['월', '화', '수', '목','금','토','일']]
y
```

```python
y.index # x에 담기
```

```python
y
```

```python
# 5. bar 차트 그리기
x = y.index # x가 없으므로 요일 정보를 출력하기 위해 인덱스명을 추출
plt.bar(x,y)

plt.ylim(500,620)

plt.xlabel('요일')
plt.ylabel('사고 건수')
plt.title('요일별 사망 교통사고 건수\n')

# ticks() : 데이터가 어디서부터 시작되는지 정확한 지점을 볼 수 있다
plt.yticks(y) # y데이터의 yticks , ylim보다 조금 더 직관적

plt.show() # show()는 항상 마지막에 써주자 
```
